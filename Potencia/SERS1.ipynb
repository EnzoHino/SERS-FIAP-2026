

# Exemplo 0 - Motor Elétrico

*1. Inserção de dados pelo usuário, com base nas especificações do motor*

PU -> Potência útil, em Watts

n -> Rendimento de 0 a 1

**n = PU / P**

FP -> fator de potência de 0 a 1

**FP = P/S**

```python
# Potencia útil (Mecanica)
pu = int(input('Insira a potência útil em KW: ')) * 1000
```

```python
# Rendimento do motor(n)
n = float(input('Insira o rendimento (de 0 a 1): '))
```

```python
# Fator de potência ( FP ou cosφ)
fp = float(input('Insira o fator de potência (0 a 1): '))
```

```python
# Cálculo da potência ativa (em Watts) -- > P
# P = PU / n

p = pu / n
print(f'A potência ativa do motor será de {int(p)} W')
```

```python
# Cálculo da potència aparente (S)
# FP = P / S
# S = P / FP
s = p / fp

print(f'A potência aparente do motor será de {int(s)} Volt-Ampére (VA)')
```

```python
# Cálculo da potència reativa (Q)
# obs.:
# **0.5 -> raiz quadrada
# **    -> potência
q = (s**2 - p**2)**0.5

print(f'A potência reativa do motor será de {q:.2f} VAr')
```

# Funções

```python
def calc_potencia_ativa(pu, n):
    p = pu / n
    return p

def calc_potencia_aparente(p, fp):
    s = p / fp
    return s

def calc_potencia_reativa(s, p):
    q = (s**2 - p**2)**0.5
    return q

def calc_triangulo_potencias(pu: float, n: float, fp: float):
  p = calc_potencia_ativa(pu, n)
  s = calc_potencia_aparente(p, fp)
  q = calc_potencia_reativa(s, p)

  print(f'A potência ativa do motor será de {int(p)} W')
  print(f'A potência aparente do motor será de {int(s)} Volt-Ampére (VA)')
  print(f'A potência reativa do motor será de {q:.2f} VAr')
```
